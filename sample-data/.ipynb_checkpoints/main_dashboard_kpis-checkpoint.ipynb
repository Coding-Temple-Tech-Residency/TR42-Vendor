```python
def unassigned_work_orders(vendor_id):
    # df = pd.read_sql(
    #     f"""
    #     select count(*) as unassigned_work_orders
    #     from work_orders
    #     where assigned_vendor='{vendor_id}'"""
    # , con=conn)
    df = pd.read_sql(
        f"""
        select work_order_id, count(*) as "unassigned tickets"
        from ticket t
        join work_orders wo
        using (work_order_id)
        where t.status = 'UNASSIGNED' and wo.assigned_vendor = '{vendor_id}'
        group by 1;
        """, con=conn
    )
    return {'vendor_id': vendor_id, 'unassigned tickets': df['unassigned tickets'].sum()}
```

```python
unassigned_work_orders(test_vendor)
```

```python
def completed_work_orders_30_days(vendor_id):
    query = """
        SELECT *
        FROM work_orders
        WHERE assigned_vendor = %(vendor_id)s
          AND current_status = 'COMPLETED'
          AND completed_at > NOW() - INTERVAL '30 days'
    """

    df = pd.read_sql(query, con=conn, params={"vendor_id": vendor_id})
    df['completion_week'] = df['completed_at'].dt.to_period('W').apply(lambda r: r.end_time).dt.date
    completed_by_week = df.groupby('completion_week')['work_order_id'].count()
    return completed_by_week
```

```python
completed = completed_work_orders_30_days(test_vendor)
completed
```

```python
def assigned_work_orders_30_days(vendor_id):
    query = """
    select * from work_orders
    where assigned_vendor = %(vendor_id)s
    and current_status != 'COMPLETED'
    and assigned_at > NOW() - interval '30 days'
    """
    df = pd.read_sql(query, con=conn, params={"vendor_id": vendor_id})
    df['assigned_week'] = df['assigned_at'].dt.to_period('W').apply(lambda r: r.end_time).dt.date
    assigned_by_week = df.groupby('assigned_week')['work_order_id'].count()
    return assigned_by_week
```

```python
assigned = assigned_work_orders_30_days(test_vendor)
assigned
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
all_weeks = pd.date_range(end=pd.Timestamp.today(), periods=5, freq='W-SUN')
```

```python
df_diff = pd.DataFrame({
    "completed": completed,
    "assigned": assigned
}).fillna(0)
```

```python
df_diff
```

```python
df_diff = df_diff.reindex(all_weeks.date, fill_value=0)
```

```python
plt.figure()

plt.plot(df_diff.index, df_diff["completed"], marker='o', label="Completed")
plt.plot(df_diff.index, df_diff["assigned"], marker='o', label="Assigned")

plt.xlabel("Week")
plt.ylabel("Number of Work Orders")
plt.title("Work Orders (Last 30 Days)")
plt.legend()

plt.gca().xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.SU))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def pending_tickets(vendor_id):
    df = pd.read_sql(f"""
    select count(*) as active_tickets
    from ticket 
    where vendor_id='{vendor_id}' and (status = 'IN_PROGRESS' or status = 'ASSIGNED')
    """, con=conn)
    count = df['active_tickets'].iloc[0]
    return {'vendor_id': vendor_id, 'active tickets': count}
```

```python
pending_tickets(test_vendor)
```
