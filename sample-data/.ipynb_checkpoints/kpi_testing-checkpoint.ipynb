# Vendor/Client Admin KPIs

## On Time Delivery %
percentage of tickets completed on time within a specified range


```python
cols = ['ticket_id', 'created_at', 'assigned_at', 'start_time', 'completed_at', 'due_date']
```

```python
def vendor_on_time_delivery(start_time, end_time=datetime.now()):
    df = tickets[tickets['created_at'].between(start_time, end_time)]
    completed_late = df['completed_at'] > df['due_date']
    overdue = (df['completed_at'].isna()) & (df['due_date'] < end_time)

    df['on_time'] = ~(completed_late | overdue)

    ticket_kpi = df.groupby('vendor_id')['on_time'].mean().round(4) * 100
    vendor_info = vendors[['vendor_id', 'company_name']]

    vendor_kpi = pd.merge(vendor_info, ticket_kpi, on='vendor_id')
    return vendor_kpi
```

```python
three_months = pd.Timestamp.now() - pd.DateOffset(months=3)
tickets_created_3_months = tickets[tickets['created_at'] >= three_months].reset_index()
```

```python
completed_late = tickets_created_3_months['completed_at'] > tickets_created_3_months['due_date']
overdue = (tickets_created_3_months['completed_at'].isna()) & (tickets_created_3_months['due_date'] < pd.Timestamp.now())

tickets_created_3_months['on_time'] = ~(completed_late | overdue)
```
