## On Time Delivery %
percentage of tickets completed on time within a specified range


```python
cols = ['ticket_id', 'created_at', 'assigned_at', 'start_time', 'completed_at', 'due_date']
```

```python
def overall_vendor_on_time_delivery(start_time, end_time=datetime.now()):
    df = tickets[tickets['created_at'].between(start_time, end_time)]
    completed_late = df['completed_at'] > df['due_date']
    overdue = (df['completed_at'].isna()) & (df['due_date'] < end_time)

    df['on_time'] = ~(completed_late | overdue)

    ticket_kpi = df.groupby('vendor_id').agg({
        'on_time': 'mean',
        'ticket_id': 'count'
    })
    ticket_kpi['on_time'] = (ticket_kpi['on_time'] * 100).round(2)
    vendor_info = vendors[['vendor_id', 'company_name']]

    vendor_kpi = pd.merge(vendor_info, ticket_kpi, on='vendor_id')
    return vendor_kpi
```

```python
def overall_vendor_ticket_duration(start_time, end_time=datetime.now()):
    df = tickets[tickets['completed_at'].between(start_time, end_time)] # all completed tickets in time range
    df['completion_time'] = df['completed_at'] - df['assigned_at']

    kpi = df.groupby('vendor_id')['completion_time'].mean().sort_values(ascending=True)
    vendor_info = vendors[['vendor_id', 'company_name']]

    vendor_kpi = pd.merge(vendor_info, kpi, on='vendor_id')

    return vendor_kpi
```

## Currently Open Tickets

```python
def overall_vendor_open_tickets():
    open_tickets = tickets[tickets['status'] != 'completed']
    kpi = pd.pivot_table(open_tickets, index='vendor_id', columns='status', values='ticket_id', aggfunc='count').reset_index()
    kpi['total'] = kpi['assigned'] + kpi['in progress']
    vendor_info = vendors[['vendor_id', 'company_name']]

    vendor_kpi = pd.merge(vendor_info, kpi, on='vendor_id')
    return vendor_kpi
```
