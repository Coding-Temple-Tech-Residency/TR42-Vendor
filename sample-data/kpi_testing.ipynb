## On Time Delivery %
percentage of tickets completed on time within a specified range


```python
cols = ['ticket_id', 'created_at', 'assigned_at', 'start_time', 'completed_at', 'due_date']
```

```python
def overall_vendor_on_time_delivery(start_time, end_time=datetime.now()):
    start_time = pd.to_datetime(start_time)
    end_time = pd.to_datetime(end_time)
    df = tickets[tickets['created_at'].between(start_time, end_time)]
    completed_late = df['completed_at'] > df['due_date']
    overdue = (df['completed_at'].isna()) & (df['due_date'] < end_time)

    df['on_time'] = ~(completed_late | overdue)

    ticket_kpi = df.groupby('vendor_id').agg({
        'on_time': 'mean',
        'ticket_id': 'count'
    })
    ticket_kpi['on_time'] = (ticket_kpi['on_time'] * 100).round(2)
    vendor_info = vendors[['vendor_id', 'company_name']]

    vendor_kpi = pd.merge(vendor_info, ticket_kpi, on='vendor_id').sort_values(by='on_time', ascending=False)
    return vendor_kpi
```

```python
def overall_vendor_on_time_delivery_2(start_time, end_time=datetime.now()):
    query = """
        select 
            v.company_name,
            round(avg(
                case 
                    when (t.completed_at > t.due_date) 
                         or (t.completed_at is null and t.due_date < ?)
                    then 0
                    else 1
                end
            ) * 100, 2) as on_time,
            count(t.ticket_id) as ticket_count
        from ticket t
        join vendor v using (vendor_id)
        where t.created_at between ? and ?
        group by v.vendor_id, v.company_name
        order by on_time desc;
    """

    vendor_kpi = pd.read_sql(
        query,
        conn,
        params=(end_time, start_time, end_time)
    )

    return vendor_kpi
```

```python
def overall_vendor_ticket_duration(start_time, end_time=datetime.now()):
    start_time = pd.to_datetime(start_time)
    end_time = pd.to_datetime(end_time)
    df = tickets[tickets['completed_at'].between(start_time, end_time)] # all completed tickets in time range
    df['completion_time'] = df['completed_at'] - df['assigned_at']

    kpi = df.groupby('vendor_id')['completion_time'].mean().sort_values(ascending=True)
    vendor_info = vendors[['vendor_id', 'company_name']]

    vendor_kpi = pd.merge(vendor_info, kpi, on='vendor_id')

    return vendor_kpi
```

## Currently Open Tickets

```python
def overall_vendor_open_tickets():
    open_tickets = tickets[tickets['status'] != 'completed']
    kpi = pd.pivot_table(open_tickets, index='vendor_id', columns='status', values='ticket_id', aggfunc='count').reset_index()
    kpi['total'] = kpi['assigned'] + kpi['in progress']
    vendor_info = vendors[['vendor_id', 'company_name']]

    vendor_kpi = pd.merge(vendor_info, kpi, on='vendor_id')
    return vendor_kpi
```

## Average Rating Per Vendor

```python
def overall_vendor_performance():
    vendor_contractor = pd.merge(vendors, contractors, on='vendor_id')
    vc_performance = pd.merge(vendor_contractor, contractor_performance, on='contractor_id')
    return vc_performance.groupby('company_name')['rating'].mean().round(2).sort_values(ascending=False)
```

```python
overall_vendor_performance()
```

```python
def overall_vendor_performance_2():
    vendor_contractor_performance = pd.read_sql_query(
        """
        select v.company_name, round(cp.avg_rating, 2) as rating
        from (
            select c.vendor_id, avg(cp.rating) as avg_rating
            from contractor_performance cp
            join contractors c using (contractor_id)
            group by c.vendor_id
        ) cp
        join vendor v using (vendor_id)
        order by cp.avg_rating desc;
        """, conn
    )
    # print(vendor_contractor_performance.head())
    return vendor_contractor_performance # vendor_contractor_performance.groupby('company_name')['rating'].mean().round(2).sort_values(ascending=False)
```
