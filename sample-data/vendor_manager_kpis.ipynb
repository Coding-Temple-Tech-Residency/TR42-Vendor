## Vendor Rating

```python
def vendor_rating(vendor_id):
    vendor_contractors = contractors[contractors['vendor_id'] == vendor_id]['contractor_id']
    df = contractor_performance[contractor_performance['contractor_id'].isin(vendor_contractors)]
    return {'vendor_id': vendor_id, 'average_rating': df['rating'].mean().round(2)}
```

## Number of active contractors

```python
def active_contractors(vendor_id):
    active_contractors = contractors[(contractors['vendor_id'] == vendor_id) & (contractors['status'] == 'active')]
    return {'vendor_id': vendor_id, 'active_contractors_num': active_contractors.shape[0]}
```

## Jobs Completed This/Last Month (work in progress)

```python
def jobs_completed_current_month(vendor_id):
    now = pd.Timestamp.now()
    start_of_month = now.replace(day=1, hour=0, minute=0, second=0, microsecond=0)

    completed_tickets_this_month = tickets[(tickets['vendor_id'] == vendor_id) & (tickets['status'] == 'completed') & (tickets['completed_at'] >= start_of_month)]
    return completed_tickets_this_month
```

```python
jobs_completed_current_month(test_vendor)
```

```python
def jobs_completed_last_month(vendor_id):
    now = pd.Timestamp.now()
    start_of_last_month = now.replace(day=1, hour=0, minute=0, second=0, microsecond=0) - pd.offsets.MonthBegin(1)

    completed_tickets_last_month = tickets[(tickets['vendor_id'] == vendor_id) & (tickets['status'] == 'completed') & (tickets['completed_at'] >= start_of_last_month)]
    return completed_tickets_last_month
```

```python
def contractor_on_time_delivery(vendor_id, start_time, end_time=None):
    if end_time is None:
        end_time = datetime.now()

    start_time = pd.to_datetime(start_time)
    end_time = pd.to_datetime(end_time)

    df = tickets[
        (tickets["vendor_id"] == vendor_id) &
        (tickets["due_date"].between(start_time, end_time, inclusive="both"))
    ].copy()

    df["on_time"] = (
        df["completed_at"].notna() &
        (df["completed_at"] <= df["due_date"])
    )

    ticket_kpi = (
        df.groupby("assigned_contractor", as_index=False)
          .agg(
              on_time_count=("on_time", "sum"),
              ticket_count=("ticket_id", "count")
          )
    )

    ticket_kpi["on_time_pct"] = (
        ticket_kpi["on_time_count"] / ticket_kpi["ticket_count"] * 100
    ).round(2)

    contractor_info = contractors[
        contractors["contractor_id"].isin(ticket_kpi["assigned_contractor"])
    ][["contractor_id", "employee_number"]]

    return (
        contractor_info.merge(
            ticket_kpi,
            left_on="contractor_id",
            right_on="assigned_contractor",
            how="inner"
        )
        .sort_values(by="on_time_pct", ascending=False)
        [["employee_number", "on_time_count", "ticket_count", "on_time_pct"]]
    )
```
